```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

## Plot data

```python
# path of the data
pth_data = 'data/'
```

```python
# load the data
signal1 = np.load(os.path.join(pth_data, 'time_series1.npy'))
signal2 = np.load(os.path.join(pth_data, 'time_series2.npy'))
```

```python
# plot the data
for i in range(len(signal1)):
    plt.plot(signal1[i], c = 'r')

for i in range(len(signal2)):
    plt.plot(signal2[i], c = 'b')

plt.show()
```
